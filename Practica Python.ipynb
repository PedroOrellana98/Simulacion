**1. Generar 2 n√∫meros pseudoaleatorios y realizar una sumatoria de los mismo**

```python
import random as ran
import matplotlib.pyplot as plot

vector= []

def repeticionCien():
    
    for i in range(100):
        dado1 =  ran.randint(1,6)
        dado2 =  ran.randint(1,6)
        tot = dado1 + dado2
        vector.append(tot)
    ##print(vector)

    intervalos = range(min(vector), max(vector) + 2)

    plot.hist(x=vector, bins=intervalos, color='#F2AB6D', rwidth=0.85)
    plot.title('Histograma 100 veces')
    plot.xlabel('Valores')
    plot.ylabel('Frecuencia')
    plot.xticks(intervalos)

    plot.show() #dibujamos el histograma

def repeticionMil():
    
    for i in range(1000):
        dado1 =  ran.randint(1,6)
        dado2 =  ran.randint(1,6)
        tot = dado1 + dado2
        vector.append(tot)
    ##print(vector)

    intervalos = range(min(vector), max(vector) + 2)

    plot.hist(x=vector, bins=intervalos, color='#F2AB6D', rwidth=0.85)
    plot.title('Histograma 1000 veces')
    plot.xlabel('Valores')
    plot.ylabel('Frecuencia')
    plot.xticks(intervalos)

    plot.show() #dibujamos el histograma
    
def repeticionDiezmil():
    
    for i in range(10000):
        dado1 =  ran.randint(1,6)
        dado2 =  ran.randint(1,6)
        tot = dado1 + dado2
        vector.append(tot)
    ##print(vector)

    intervalos = range(min(vector), max(vector) + 2)

    plot.hist(x=vector, bins=intervalos, color='#F2AB6D', rwidth=0.85)
    plot.title('Histograma 10000 veces')
    plot.xlabel('Valores')
    plot.ylabel('Frecuencia')
    plot.xticks(intervalos)

    plot.show() #dibujamos el histograma

repeticionCien() 
repeticionMil()
repeticionDiezmil()
```
