```python
from os import sep
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

directorio1 = "/content/drive/MyDrive/Simulacion/Libro1.xlsx"
df1 = pd.read_excel(directorio1)
df1
```

```python
df1["Variación Ranking Tamaño 2020 - 2021"]
variacion = df1.loc[:,"Variación Ranking Tamaño 2020 - 2021"]
np=variacion.values
menores=[]
mayores=[]

for i in variacion:
    if i < 0:
      menores.append(i)
    if i > 1:
      mayores.append(i)

print(len(menores))
print(len(mayores))
total = [len(menores), len(mayores)]

vector = ['Empeora', 'Mejora']
## Creamos Gráfica
plt.bar(vector, total)
## Legenda en el eje y
plt.ylabel('# Empresas')
## Legenda en el eje x
plt.xlabel('\nTipo de variacion')
## Título de Gráfica
plt.title('Variación Ranking Tamaño 2020 - 2021')
## Mostramos Gráfica
plt.show() 
```

```python
df1["No. de Integrantes y Empresas Offshore relacionadas con el Grupo Económico publicados en PANAMÁ PAPERS (***)"]
papers = df1.loc[:,"No. de Integrantes y Empresas Offshore relacionadas con el Grupo Económico publicados en PANAMÁ PAPERS (***)"]
np=papers.values
contPapers=[]

for i in papers:
    if i > 0:
      contPapers.append(i)

print(len(contPapers))
#print(contPapers)

vector = ['Contiene']
## Creamos Gráfica
plt.bar(vector, len(contPapers))
## Legenda en el eje y
plt.ylabel('# Papers')
## Título de Gráfica
plt.title('No. de Integrantes y Empresas Offshore relacionadas con el Grupo Económico publicados en PANAMÁ PAPERS')
## Mostramos Gráfica
plt.show() 
```

```python
df2 = df1.sort_values(["Contribución   I.R. 2020"], ascending=False)
ordenado = df2["Contribución   I.R. 2020"]
grupo = df2["\nNombre Grupo Económico"]
ordenado = ordenado*100
top10 = [ordenado[0:10], grupo[0:10]]

plt.figure()

colores = ['#abcdef', '#aabbcc', '#abcabc', '#ccbbaa']
plt.pie([ordenado[0:10]], labels=grupo[0:10], colors=colores, autopct='%.2f %%')
plt.title('Empresas que mas aportaron')
plt.show()
```

```python
df3 = df1.sort_values(["Total Patrimonio (7) 2020"], ascending=False)
ordenado = df3["Total Patrimonio (7) 2020"]
grupo = df3["\nNombre Grupo Económico"]
top10 = [ordenado[1:11], grupo[1:11]]
#print(top10)

plt.bar(grupo[1:11], ordenado[1:11])
plt.xticks(rotation=90)
## Legenda en el eje y
plt.xlabel('Nombre Grupo Económico')
## Legenda en el eje y
plt.ylabel('Total patrimonio')
## Título de Gráfica
plt.title('Top 10 mayor patrimonio 2020')
## Mostramos Gráfica
plt.show() 
```
