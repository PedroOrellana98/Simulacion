**1 Con los datos de Fifa, organizar a los jugadores de acuerdo al peso en las siguientes escalas y generar un cuadro tipo PIE**

* Debajo 125 Lbs.
* 125-150
* 150-175
* 175 o superior



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

directorio = "/content/drive/MyDrive/Simulacion/Fifa/fifa_datos.csv"
df = pd.read_csv(directorio)
df['Weight'] = df['Weight'].str.replace('lbs','')
peso = df.loc[:,"Weight"]
np=peso.values
#print(np)

list1 = []
list2 = []
list3 = []
list4 = []
vector0 = []
vector1 = []
vector2 = []
vector3 = []

for i in range(125, 151):
    list1.append(str(i).zfill(1))

for i in peso:
  for j in list1:
    if i == j:
      vector0.append(i)

print(len(vector0))

for i in range(0, 125):
    list2.append(str(i).zfill(1))

for i in peso:
  for j in list2:
    if i == j:
      vector1.append(i)

print(len(vector1))

for i in range(150, 176):
    list3.append(str(i).zfill(1))

for i in peso:
  for j in list3:
    if i == j:
      vector2.append(i)

print(len(vector2))

for i in range(176, 300):
    list4.append(str(i).zfill(1))

for i in peso:
  for j in list4:
    if i == j:
      vector3.append(i)

print(len(vector3))

plt.figure()

etiquetas = ['Peso Medio 1', 'Peso Bajo', 'Peso Medio2', 'Peso Alto']
colores = ['#abcdef', '#aabbcc', '#abcabc', '#ccbbaa']
plt.pie([len(vector0), len(vector1), len(vector2), len(vector3)], labels=etiquetas, colors=colores, autopct='%.2f %%')
plt.title('Escalas de dataset Fifa')
plt.show()
```

**2 Generar un grafico de barras (histograma) de acuerdo a su habilidad (Overall)  en base a los siguientes segmentos contando el n√∫mero de jugadores**

* 40
* 50
* 60
* 70
* 80
* 90

```python
import matplotlib.pyplot as plot
general = df.loc[:,"Overall"]
np=general.values
cuarenta = []
cincuenta = []
sesenta = []
setenta = []
ochenta = []
noventa = []

for i in general:
    if i == 40:
      cuarenta.append(i)
    if i == 50:
      cincuenta.append(i) 
    if i == 60:
      sesenta.append(i) 
    if i == 70:
      setenta.append(i) 
    if i == 80:
      ochenta.append(i) 
    if i == 90:
      noventa.append(i)   

print("Vector de 40 = " + str(len(cuarenta)) 
+ "\nVector de 50 = " + str(len(cincuenta))
+ "\nVector de 60 = " + str(len(sesenta))
+ "\nVector de 70 = " + str(len(setenta))
+ "\nVector de 80 = " + str(len(ochenta))
+ "\nVector de 90 = " + str(len(noventa)))

vec = [cuarenta,cincuenta,sesenta,setenta,ochenta,noventa]
valores = [40, 50, 60, 70, 80, 90]
colore = ['#abcdef', '#aabbcc', '#abcabc', '#ccbbaa', '#bbbbaa', '#ccccbb']

plot.hist(x=vec, bins=valores, color=colore)
plot.title('Histograma Overall')
plot.xlabel('General')
plot.ylabel('Frecuencia')
plot.xticks(valores)

plot.show() #dibujamos el histograma
```
