**Cuadrados Medios**

```python
xo = int(input("Ingrese la semilla: "))
digitos = int(input("Ingrese el numero de digitos: "))
iteraciones = int(input("Ingrese el numero de iteraciones: "))
xn = xo
print("It. \t Xn \t Ui \t Rn")
vectorXn =[]

for i in range(iteraciones):
  xnn = xn**2
  txnn = str(xnn).zfill(8)
  tam = len(txnn)
  ui = int(txnn[int(tam/2-digitos/2):int(tam/2+digitos/2)])
  rn = ui / (int('9'*digitos)+1)
  print(str(i) + "\t" + str(xn) + "\t"+ str(ui)+"\t"+str(rn))
  xn = ui
  vectorXn.append(rn)
```

**Congruencia Lineal**

```python
xo = int(input("Ingrese el valor de Xo: "))
a = int(input("Ingrese el valor a: "))
b = int(input("Ingrese el valor b: "))
m = int(input("Ingrese el valor m: "))
iteraciones = int(input("Ingrese el numero de iteraciones: "))
xn = xo
vectorXn1 = []

print("It. \t Xn \t Ui")
for i in range(iteraciones):
  xnn = (a*xn+b) % m
  ui = xnn/m
  print(str(i) + "\t"+ str(xnn)+"\t"+str(ui))
  xn = xnn
  vectorXn1.append(ui)
```

**Tausworthe**

```python
from itertools import zip_longest

r = int(input("valor de r: "))
q = int(input("valor de q: "))
baseb = int(input("valor de base binaria: "))
vectorBase =[]

if r < q :
  def operacionXOR(a,b):
    ab = 1
    if a  == b:
      ab = 0
    return ab

  bits = []
  b = (2**q) - 1

  for i in range(0,b):
    bits.append(0)

  for i in range(0,q):
    bits[i] = 1
    bits.append(1)

  a = q+1
  for i in range(a,len(bits)):
    i1 = i - r
    i2 = i - q
    bits[i] = operacionXOR(bits[i1],bits[i2])


  def binarioADecimal(binario):
    a = 0
    p = (baseb-1)
    for i in range(0,len(binario)):
      if binario[i] == 1:
        a += 2**(p-i)
    return a
  test_list = bits

  def elementos(n, iterable, padvalue='1'):
      return zip_longest(*[iter(iterable)]*n, fillvalue=padvalue)

  print("\n","It.", "\t", "Base 2", "\t","Base 10","\t","Ui","\n")
  d = 0
  for output in elementos(baseb,test_list):
      vectorBase.append(binarioADecimal(output)/(2**baseb))
      lst_new = [str(a) for a in output]
      print(d,"\t" ," ".join(lst_new), "\t", binarioADecimal(output), "\t", "\t", binarioADecimal(output)/(2**baseb))
      d +=1
else :
  print("El valor de r=" + str(r) + " tiene que ser menor a q=" + str(q))
```

```python
# Importanmos las librerias
import random
import matplotlib.pyplot as plt
import numpy as np
import math

def random_python(N=100):
  numeros = []
  random.seed(1032022)
  [numeros.append(random.random())for i in range(N)]
  return numeros

def conteoCorridas(numeros):  
  num = []
  for i in range(len(numeros)):
    if numeros[i] >= 0.5:
      num.append(1)
    else:
      num.append(0)

  puntero = num[0]
  contador = 0
  for i in range(len(num)):
    if num[i] != puntero:
      contador += 1
    puntero = num[i]
  return contador

def conteoCero(numeros):  
  num = []
  for i in range(len(numeros)):
    if numeros[i] >= 0.5:
      num.append(1)
    else:
      num.append(0)
   
  contador = 0
  for i in range(len(num)):
    if num[i] == 0:
      contador += 1  
  return contador  

def conteoUno(numeros):  
  num = []
  for i in range(len(numeros)):
    if numeros[i] >= 0.5:
      num.append(1)
    else:
      num.append(0)
   
  contador = 0
  for i in range(len(num)):
    if num[i] == 1:
      contador += 1  
  return contador   


# Listo eso es todo
numeros = random_python()
print("Corrida Semilla \n\nLos 100 numeros aleatorios generados\n")
print(numeros)
print("\nNumeros Generados\n")
tablas = conteoCorridas(numeros)
tablas1 = conteoCero(numeros)
tablas2 = conteoUno(numeros)
print("Corridas "+str(tablas+1))
print("Ceros "+str(tablas1))
print("Unos "+str(tablas2))
uc=((2*tablas1*tablas2)/(tablas1+tablas2))+0.5
print("Uc "+str(uc))
aC=(((2*tablas1*tablas2)*(2*tablas1*tablas2-(tablas1+tablas2))))
ac0=((tablas1+tablas2)**2)*((tablas1+tablas2)-1)
ac1=aC/ac0

print("Varianza "+str(ac1))

est = ((tablas) - uc) / math.sqrt(ac1)
print("ESTADISTICO Z0: ", est)

vector1=[tablas,tablas1,tablas2]
print("Grafica de barras")
vector=["Corridas","Ceros","Unos"]
colores = ['#abcdef', '#aabbcc', '#abcabc', '#ccbbaa']
plt.pie([vector1], labels=vector, colors=colores, autopct='%.2f %%')
plt.title('Corridas Semilla')
plt.show()
```
